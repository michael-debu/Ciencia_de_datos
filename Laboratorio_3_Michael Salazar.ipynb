##### 1. Cree un procedimiento en Python dado un numero entero en decimal despliegue su representacion en numero binario.

```python
def decimal_a_binario(n):
    binario = ""
    
    while n > 0:
        residuo = n % 2
        binario = str(residuo) + binario
        n = n // 2
    
    return binario

print(decimal_a_binario(124445))
```

##### 3. Haga un programa en Python que dada una lista de n numeros, calcule la desviacion estandar del conjunto de numeros, para este ejercicio no puede utilizar ninguna funcion que resuelva el problema.

```python
def desviacion(lista):
    suma = 0
    
    for x in lista:
        suma = suma + x
    
    promedio = suma / len(lista)
    
    suma2 = 0
    for x in lista:
        suma2 = suma2 + (x - promedio) ** 2
    
    varianza = suma2 / len(lista)
    
    return varianza ** 0.5

print(desviacion([10, 20, 30, 40,100,21,45,65,98,65,34,62,74,90,1,89]))
```

##### 4.Desarrolle en Python una funcion que dado un arreglo de corra el algoritmo de ordenamiento burbuja y devuelva un nuevo arreglo ordenado. Posteriormente cree una funcion main para probar su funcion.

```python
def ordenar(numeros):
    for i in range(len(numeros)):
        for j in range(len(numeros) - 1):
            if numeros[j] > numeros[j + 1]:
                numeros[j], numeros[j + 1] = numeros[j + 1], numeros[j]
    return numeros

print(ordenar([5, 2, 9, 1,9,8,10]))
```

#### 6. Cree un programa en Python que dada una lista de tuplas, elimine las tuplas que est´an vac´ıas.

* Si la lista fuera: [(), (), (’X’,), (’a’, ’b’), (’a’, ’b’, ’c’), (’d’)]

```python
def eliminar_vacias(lista):
    nueva = []
    
    for x in lista:
        if x != ():
            nueva.append(x)
    
    return nueva

lista = [(), (), ('Y',),(),(), ('d', 'f'),(),(), ('e', 'b', 'c'),(),(),('d',)]
print(eliminar_vacias(lista))
```

#### 7. Cree una funcion en Python que dada una tupla de tuplas con numeros, produzca una tupla con el promedio de cada tupla individual. Tomar nota del siguiente ejemplo:
* Si la tupla de entrada fuera: ((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 32))
* El resultado deberıa ser: (10.5, 44, 58)

```python
def promedio_tuplas(tuplas):
    resultado = []
    
    for t in tuplas:
        suma = 0
        
        for x in t:
            suma = suma + x
        
        promedio = suma / len(t)
        resultado.append(promedio)
    
    return tuple(resultado)


datos = ((20, 30, 90, 14), (14, 23, 90, 100), (52, 87, 98, 23))
print(promedio_tuplas(datos))
```

#### 8. El objetivo es crear una representacion de un libro usando programacion orientada a objetos. Debes definir una clase llamada Libro que tenga atributos como el tıtulo, el autor, el numero de paginas y un indicador de si el libro ha sido le´ıdo o no. Ademas, debes implementar metodos para: 
* Inicializar un nuevo libro con su informacion.
* Marcar un libro como leıdo.
* Producir una descripcion legible del libro que incluya todos sus detalles

```python
class Libro:
    def __init__(self, titulo, autor, paginas):
        self.titulo = titulo
        self.autor = autor
        self.paginas = paginas
        self.leido = False

    def marcar_leido(self):
        self.leido = True

    def descripcion(self):
        return self.titulo + " " + self.autor+ " " + str(self.paginas) + "  páginas. Estatus de lectura: " + str(self.leido)

## libro esta como false por que al principio no esta leido 
libro1 = Libro("La astronomía en 100 preguntas", "Vicent Martínez Badenes", 1000)

print(libro1.descripcion())
```

```python
## cambio de estatus de lectura 
libro1.marcar_leido()

print(libro1.descripcion())
```

#### 9. Cree una clase llamada Circulo con un atributo para el radio del cırculo. Luego, implementaras metodos para:
* Inicializar un nuevo cırculo con su radio.
* Calcular el area del cırculo.
* Calcular la circunferencia del cırculo.

```python
class Circulo:
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return round(( 3.1416 * self.radio * self.radio),2)

    def circunferencia(self):
        return round ((2 * 3.1416 * self.radio),2)

c = Circulo(6371)

print("Área:", c.area())
print("Circunferencia:", c.circunferencia())
```
